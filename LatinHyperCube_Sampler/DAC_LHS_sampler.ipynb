```python
#Importing the required library for latin hypersquare sampling
from scipy.stats import qmc
import pandas as pd
import csv
```

```python
#Define the variables and acceptable ranges [cac03_flow_rate, calciner_temp, pellet_moisture_frac, air_mol_flow_rate, RH_air, DAC_column_height, pellet_diameter, DAC_ambient_temp]
lhs_variables =  {
    "caco3_flow_rate [kmol/hr]": [5, 100],                       # in kmol/hr
    "calciner_temp [Deg C]": [890, 950],                            #calciner temp without too much sintering (in degree c)
    "pellet_moisture_frac": [0.05, 0.5],                            # a decimal number
    "air_mol_flow_rate [kmol/hr]": [7000, 9000],                    # in kmol/hr
    "RH_air": [0.1, 0.9],                                           # relative humidity, less than 1
    "DAC_column_height [m]": [1.0, 10.0],                            # height in meters
    "pellet_diameter [mm]": [1.0, 6.0],                             # in mm
    "DAC_ambient_temp [Deg C]": [15, 45],                           # in degree celsius
    "e_voidage": [0.35, 0.45]                                         #Voidage in the DAC unit
}


#Set your dimensions d and initiate sampler
sampler = qmc.LatinHypercube(d=len(lhs_variables), seed=42)

#Next, set the number of samples preferred
sample = sampler.random(n=1100)

#scale samples to specific ranges, important as industrial processes have real constraints. format [min, max]. 
#[cac03_flow_rate, calciner_temp, pellet_moisture_frac, air_flow_rate, RH_air, DAC_column_height, pellet_diameter, DAC_temp]
#define the lower and upper bounds
l_bounds = [r[0] for r in lhs_variables.values()]
u_bounds = [r[1] for r in lhs_variables.values()]

#generate and print scaled sample

scaled_sample = qmc.scale(sample, l_bounds, u_bounds)

#convert to data frame
df_lhs = pd.DataFrame(scaled_sample, columns=lhs_variables.keys())
index = [i for i in range(0, len(df_lhs))]
df_lhs["sim_index"] = index

#export to csv
df_lhs.to_csv("LHS_variables_to_be_manipulated_by_python.csv", index=False, encoding="utf-8")

```
